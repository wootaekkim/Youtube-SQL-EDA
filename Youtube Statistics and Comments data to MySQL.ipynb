create table in mysql for comments first

```python
cursor = conn.cursor()
```

```python
cursor.execute("USE youtube;")
```

```python
cursor.execute("DROP TABLE YoutubeStats;")
```

```python
try:
    cursor.execute("CREATE TABLE Comments(VideoID varchar(255), Comment LONGTEXT, Likes DECIMAL(14, 1), Sentiment INT)")
    print("Table is created....")
    for i,row in comm_df.iterrows():
            #here %S means string values 
            sql = "INSERT INTO youtube.Comments VALUES (%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not auto committed by default, so we must commit to save our changes
            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```

Now create table for Youtube statistics

```python
try:
    cursor.execute("CREATE TABLE YoutubeStats(Title LONGTEXT, VideoID varchar(255), PublishedAt DATE, Keyword varchar(255), Likes DECIMAL(14, 1), Comment DECIMAL(14, 1), Views DECIMAL(14, 1))")
    print("Table is created....")
    for i,row in vidstat_df.iterrows():
            #here %S means string values 
            sql = "INSERT INTO youtube.YoutubeStats VALUES (%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not auto committed by default, so we must commit to save our changes
            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```
